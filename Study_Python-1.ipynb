## Plots

```python
# Scatter Plot
fig, (ax,ax2) = plt.subplots(1,2,figsize=(16,5))
colors = {'setosa': 'r', 'versicolor': 'g', 'virginica': 'b'}
for i in range(len(iris)):
    ax.scatter(iris['petal_length'][i], iris['petal_width'][i], color=colors[iris['species'][i]])
    ax2.scatter(iris['sepal_length'][i], iris['sepal_width'][i], color=colors[iris['species'][i]])
ax.set_title('Iris Dataset')
ax.set_xlabel('petal_length')
ax.set_ylabel('petal_width')
ax2.set_title('Iris Dataset')
ax2.set_xlabel('sepal_length')
ax2.set_ylabel('sepal_width')
```

```python
import seaborn as sns #Visulization
```

```python
#histogram
f, ax = plt.subplots(figsize=(8, 6))
sns.distplot(iris['petal_length'])
```

```python
# correlation이 높은 상위 4개의 heatmap
# continuous variables --> pearson
# continuous + sequential variables --> spearman
# abs는 반비례관계도 고려하기 위함
import scipy as sp

cor_abs = abs(iris.corr(method='spearman')) 
cor_cols = cor_abs.nlargest(n=4, columns='petal_length').index # petal_length correlation이 높은 column 4개 뽑기(내림차순)
# spearman coefficient matrix
cor = np.array(sp.stats.spearmanr(iris[cor_cols].values))[0] # 4 x 4
print(cor_cols.values)
plt.figure(figsize=(10,10))
sns.set(font_scale=1.25)
sns.heatmap(cor, fmt='.2f', annot=True, square=True , annot_kws={'size' : 8} ,xticklabels=cor_cols.values, yticklabels=cor_cols.values)
```
