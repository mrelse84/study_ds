# Study Python Basics

```python
import numpy as np
import pandas as pd
```

# Numpy

```python
# ndarray concatenate
arr1 = np.array([1,2,3,4,5])
arr2 = np.array(['a','b','c','d','e'])
arr3 = np.concatenate((arr1.reshape(-1,1),arr2.reshape(-1,1)), axis=1)
arr3
```

# Pandas

```python
map_data1 = {'a':[1,2,3,4], 'b':[5,6,7,8], 'c':[9,10,11,12], 'd':[13,14,15,16]}
df1 = pd.DataFrame(map_data1) # Same as pd.DataFrame.from_dict(map_data1)
print(df1, '\n', df1.index, '\n', df1.columns)
```

```python
map_data2 = {1:[10,20,30,40], 2:[50,60,70,80], 3:[90,100,110,120], 4:[130,140,150,160]}
df2 = pd.DataFrame.from_dict(map_data2, orient='index', columns=['a','b','c','d'])
print(df2, '\n', df2.index, '\n', df2.columns)
```

```python
# One Col
df1['a'], type(df1['a']), df1[['a']], type(df1[['a']]) # Series, DataFrame
```

```python
# Multi Cols
print(df1[['a','c']], '\n\n', df1.iloc[:,1:3])
```

```python
# One Row & Multi Rows
print(df1.iloc[1], '\n\n', type(df1.iloc[1]), df1[1:3], '\n\n', type(df1[1:3]))
#print(df2.iloc[1], '\n\n', type(df2.iloc[1]), df2[1:3], '\n\n', type(df2[1:3]))
```
